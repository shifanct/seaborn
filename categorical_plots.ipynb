# countplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
df.dtypes
```

```python
# sns.countplot(df)
```

```python
sns.countplot(x="sex",data=df)
```

## horizondal countplot

```python
sns.countplot(y='day',data=df)
```

## palette

```python
sns.countplot(x='day',data=df,palette='rocket')
```

## hue

```python
import matplotlib.pyplot as plt
sns.countplot(x="time",hue="sex",data=df)
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="time",hue="sex",data=df,dodge=False)
```

## rotating Labels

```python
import matplotlib.pyplot as plt
sns.countplot(x="time",hue="sex",data=df)
plt.xticks(rotation=30)
```

## ordedr

```python
sns.countplot(x='day',data=df,palette="magma")
```

```python
sns.countplot(x='day',data=df,palette="magma",order=['Thur', 'Sat', 'Fri','Sun',])
```

# Bar plot()

```python
df=sns.load_dataset("tips")
df
```

```python
# by default the barplot shows the mean if there multiple values
sns.barplot(x="day",y="tip",data=df)
```

# Estimator

```python
sns.barplot(x="day",y="tip",data=df,estimator="sum")
```

```python
sns.barplot(x="day",y="tip",data=df,estimator="max")
```

# hue()

```python
sns.barplot(x="sex",y="size",hue="smoker",data=df)
```

```python
sns.barplot(x="time",y="size",hue="sex",palette="magma",data=df)
```

```python
sns.barplot(x="time",y="size",hue="sex",palette="magma",dodge=False,data=df)
```

## horizondal bars

```python
plt.figure(figsize=(6,8))
sns.barplot(y="day",x="tip",palette="rocket",data=df)
```

## errorbar

```python
sns.barplot(x="day",y="tip",data=df,estimator="max",errorbar=None)
```

```python
color=["green",'green','red','green']
sns.barplot(x="day",y="tip",data=df,estimator="max",errorbar=None,palette=color)
```

# Box plot

```python
df=sns.load_dataset("tips")
df
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
```

## hue

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="smoker")
```

# Adding colors

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="smoker",palette="coolwarm")
```

## hiding outliers

```python
sns.boxplot(x="day",y="total_bill",data=df,hue="smoker",palette="coolwarm",showfliers=False)
```

# violin plot

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.violinplot(x="time",y="total_bill",data=df)
```

## Horizondal violinplot

```python
sns.violinplot(x="total_bill",y="day",data=df)
```

## Hue

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="sex")
```

## palette

```python
sns.violinplot(x="time",y="total_bill",data=df,palette="magma")
```

## split

```python
# to show the half portions of the violinplot
sns.violinplot(x="time",y="total_bill",data=df,palette="magma")
```

## cut

```python
# the sharpest edge shows the more number of outliers
sns.violinplot(x="day",y="tip",data=tips)
```

```python
# we can use the cut argument for hide the outliers on a boxplot
sns.violinplot(x="day",y="tip",data=tips,cut=0)
```
