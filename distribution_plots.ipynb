# Histogram

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.histplot(x="size",data=tips)
```

## Add KDE

```python
sns.histplot(tips["total_bill"],kde=True)
```

## customize number of bins

```python
sns.histplot(tips["total_bill"],kde=True,bins=10)
```

## hue

```python
                                                                # default multiple is layer
sns.histplot(data=tips,x="total_bill",kde=True,palette="viridis",hue="sex")
```

```python
sns.histplot(x="size",hue="sex",data=tips)
```

```python
sns.histplot(x="size",hue="sex",data=tips,multiple='dodge') # stack,
```

```python
sns.histplot(x="size",hue="sex",data=tips,multiple='stack')
```

## color & palette

```python
sns.histplot(data=tips,x="total_bill",hue="smoker",edgecolor="red",palette="rocket")
```

# Kde plot

```python
import seaborn as sns
df=sns.load_dataset("tips")
df
```

```python
sns.kdeplot(df["tip"])
```

```python
sns.kdeplot(data=df,x="size",y="tip")
```

## vertical plot

```python
sns.kdeplot(y="total_bill",data=df)
```

```python
sns.kdeplot(df["total_bill"],vertical=True)
```

## Fill

```python
sns.kdeplot(x="tip",data=df,fill=True)
```

## hue

```python
sns.kdeplot(x="total_bill",data=df,hue="sex")
```

```python
sns.kdeplot(x="total_bill",data=df,hue="day")
```

## shade

```python
sns.kdeplot(x="total_bill",data=df,hue="day",fill=True)
```

```python
sns.kdeplot(x="total_bill",data=df,hue="day",shade=True, palette='magma')
```

## color

```python
sns.kdeplot(df["size"],color="red")
```

```python
sns.kdeplot(data=df, x="total_bill", shade =True, facecolor='g')
```

# Rug plot

```python
sns.rugplot(df["total_bill"])
```

## Oriantation

```python
sns.rugplot(y=df["total_bill"])
```

## Ticks

```python
sns.rugplot(x="size",data=df,height=0.2)
```

## color

```python
sns.rugplot(x="total_bill",data=df,color="g")
```

## combine rugplot with histograms

```python
sns.histplot(x="total_bill",data=df,kde=True,color="red",edgecolor="black")
sns.rugplot(df["total_bill"],color="black",height=0.1)
```

```python
sns.kdeplot(x="total_bill",data=df,color="green",fill=True)
sns.rugplot(x="total_bill",data=df,height=0.2,color="w")
```

## Hue

```python
sns.rugplot(data=df,x="total_bill",hue="sex")
```

```python
sns.rugplot(data=df,x="total_bill",hue="sex",palette="rocket")
```

```python
sns.histplot(data=df,x='total_bill',hue='sex',kde=True)
sns.rugplot(data=df,x='total_bill',hue='sex',height=0.1)
```
