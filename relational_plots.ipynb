# Line plot

```python
import seaborn as sns
# load dataset
df=sns.load_dataset("tips")

# basic line plot
sns.lineplot(x="size",y="total_bill",data=df)
```

## remove ci

```python
sns.lineplot(x="size",y="total_bill",data=df,ci=None,color="red")
```

## hue

```python
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",ci=None)
```

```python
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",ci='sd')
```

## style

```python
color=["green","blue"]
sns.lineplot(x="size",y="total_bill",data=df,ci=None,palette=color,style="smoker")
```

```python
sns.lineplot(x="size",y="total_bill",data=df,ci=None,color="red",style="sex")
```

## size

```python
sns.lineplot(x="size",y="total_bill",data=df,ci=None,style="sex",size="time")
```

```python
sns.lineplot(x="size",y="total_bill",hue="smoker",data=df,ci=None,style="sex",size="time",sizes=(2,10))
```

```python
# specify colors
colors=["red",'green']
sns.lineplot(x="size",y="total_bill",data=df,style="time",hue="time",palette=colors)
```
