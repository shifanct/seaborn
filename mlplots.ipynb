# MinMax Scaler

```python
# importing the visualixation librery and MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

tips=sns.load_dataset('tips')
scaler_minmax = MinMaxScaler()
tips['total_bill_MinMax'] = scaler_minmax.fit_transform(tips[['total_bill']])


# plotting histogram of total_bill
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
sns.histplot(tips['total_bill'],kde=True)
plt.title("original total_bill")


plt.subplot(1,2,2)
sns.histplot(tips['total_bill_MinMax'],kde=True)
plt.title('MinMax scaled total_bill')
plt.show()
```

# Roc curve

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,auc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Train the random forest classifier
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
# predict Probabilities
y_probs=clf.predict_proba(x_test)[:,1]
```

```python
fpr,tpr,threshould=roc_curve(y_test,y_probs)
roc_auc=auc(fpr,tpr)
```

```python
# plot the roc curve
sns.lineplot(x=fpr,y=tpr,label=f'ROC curve(AUC={roc_auc:.2f}')
plt.plot([0,1],[0,1],linestyle='--',color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

# with Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.datasets import load_iris

# step 1: load the iris dataset
iris = load_iris()
x=iris.data
y=(iris.target ==2).astype('int')

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

# step 2: create a logistic regression model
model=LogisticRegression(random_state=42)

# step 3: train the model
model.fit(x_train,y_train)

y_pred=model.predict_proba(x_test)[:,1]

fpr,tpr,threshould=roc_curve(y_test,y_pred)
auc_score = roc_auc_score(y_test, y_pred)


# step 6: plot the ROC Curve using seaborn stylling
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'ROC Curve (AUC={auc_score:.2f})',color="blue",lw=2)
plt.plot([0,1],[0,1],linestyle='--',color='gray',label='Random Classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

# precision-Recall Curve

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.datasets import load_iris

# step 1: load the iris dataset and prepare binary labels
iris = load_iris()
x=iris.data
y=(iris.target ==2).astype('int')

# step 2: plit the l
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

# step 3: train the model
model=LogisticRegression(random_state=42)


model.fit(x_train,y_train)

# step4 :
y_probs=model.predict_proba(x_test)[:,1]

#step5: cumpute the precision-recall curve and average precision score
precision,recall,threshould=precision_recall_curve(y_test,y_probs)
average_precision = average_precision_score(y_test,y_probs)



# step 6: plot the ROC Curve using seaborn stylling
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,6))
plt.plot(recall,precision,label=f'precision recall Curve (ap;={average_precision:.2f})',color="blue",lw=2)
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('precision recall Curve')
plt.legend(loc='lower left')
plt.show()
```

# Residal plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

tips=sns.load_dataset('tips')

X=tips[['total_bill']]
y=tips['tip']

# train the linearregression model
model=LinearRegression()
model.fit(X,y)

# predict tips using the model
y_pred=model.predict(X)

#calculate residuals
residuals=y-y_pred

#create a residentional plot
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,6))
sns.residplot(x=y_pred,y=residuals,color="blue",lowess=True,line_kws={'color':'red'})
plt.axhline(0, linestyle='--',color='gray',linewidth=1) # Reference line at 0
plt.xlabel("Predicted Tips")
plt.ylabel("Residuals")
plt.title("Residual Plot: tips dataset (Linear Regression)")
```
